### Import Modules

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

### Load the webpage

```python
url='https://www.flipkart.com/cameras/dslr-mirrorless/pr?sid=jek,p31,trv'
r=requests.get(url)
r
```

```python
htmlcontent=r.content
htmlcontent
```

```python
soup=bs(htmlcontent,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
product_name = []
product_price = []
ratings = []
#no_reviews = []
#no_ratings = []
image_link = []
for i in range(1):
    
    print('page scrapping',i+1)
    
    Titles=soup.find_all("div",class_="_4rR01T")
    for t in Titles:
        product_name.append(t.get_text())
    
    price=soup.find_all("div",class_="_30jeq3 _1_WHN1")
    for p in price:
        product_price.append(p.get_text())
        
    stars=soup.find_all("div",class_="_3LWZlK")
    for s in stars[:24]:
        ratings.append(s.get_text())
        
    image = soup.find_all('img',class_='_396cs4')
    for i in image[:24]:
        image_link.append(i.get('src'))
        
    requests.get('https://www.flipkart.com/cameras/dslr-mirrorless/pr?sid=jek%2Cp31%2Ctrv&page={i+2}')
    
```

```python
#converting scrap into CSV

import pandas as pd
df=pd.DataFrame()
df["Product"]=product_name
df["Price"]=product_price
df["Ratings"]=ratings
df['Images'] = image_link
```
